```python
# Check country where name equal to pakistan from facts table
q = """select * from facts 
        where name == "Pakistan"

"""
sql_query(q)
```

```python
# Check city where facts_id = 132
q = """select * from cities
        where facts_id = 132

"""
sql_query(q)
```

```python
# join two table check country name and capital
q = """select f.name Country_name , c.name Capital from facts f
        inner join cities c
        on 
        f.id = c.facts_id
        where c.capital == 1
"""
sql_query(q)
```

```python
# join two table where capital == 0
q = """select f.name Country_name , c.name Capital from facts f
        inner join cities c
        on 
        f.id = c.facts_id
        where c.capital == 0
"""
sql_query(q)
```

```python
q = """select f.name Country, c.name City from facts f
        left join cities c
        on
        f.id = c.facts_id
        where c.name is null
        order by 1
"""
sql_query(q)
```

```python
# Country, Capital, Capital_population where Capital_population > 10000000

q = """select f.name Country, c.name Capital , c.population Capital_population from facts f
        inner join cities c
        on 
        f.id = c.facts_id
        where Capital_population > 10000000

"""
sql_query(q)

```

```python
# Where capital == 1 and Capital_pop > 10000000

q = """select f.name Country, c.name Capital, c.population Capital_population
       from facts f
       inner join cities c
       on
       f.id = c.facts_id
       where c.Capital == 1 and Capital_population > 10000000
       

"""
sql_query(q)
```

```python

# Where capital == 0 and Capital_pop > 10000000

q = """select f.name Country, c.name Capital, c.population Capital_population
       from facts f
       inner join cities c
       on
       f.id = c.facts_id
       where c.Capital == 0 and Capital_population > 10000000
       

"""
sql_query(q)
```

```python
# TOp 3 Capital population Where capital == 0 and Capital_pop > 10000000

q = """select f.name Country, c.name Capital, c.population Capital_population
       from facts f
       inner join cities c
       on
       f.id = c.facts_id
       where c.Capital == 0 and Capital_population > 10000000
       order by 3 DESC limit 3

"""
sql_query(q)
```

```python
# Country, Top_pop, Urban_pop, Urban_pop_pct
```

```python
q = """select f.name Country, f.population Top_pop, c.population Urban_pop from facts f
        inner join cities c
        on 
        f.id = c.facts_id
"""
sql_query(q)
```

```python
q = """select f.name Country, f.population Top_pop, sum(c.population) Urban_pop from facts f
        inner join cities c
        on 
        f.id = c.facts_id
        group by 1
"""
sql_query(q)
```

```python
q = """select f.name Country, f.population Top_pop, sum(c.population) Urban_pop,
       round(cast(c.population as float)/f.population*100, 2) Urban_pop_per from facts f
        inner join cities c
        on 
        f.id = c.facts_id
        group by 1
"""
sql_query(q)
```
