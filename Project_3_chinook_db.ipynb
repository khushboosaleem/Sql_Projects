```python
# Top five Customer_name, No_of_invoices, Total_spent

q = """select c.first_name ||" "|| c.last_name Customer_name,
        count(inv.invoice_id) No_of_Invoices,
        sum(inv.total) Total_spen
        from customer c
        inner join invoice inv
        on
        c.customer_id = inv.customer_id
        group by c.customer_id
        order by 3 DESC limit 5

"""
sql_query(q)
```

```python
# Add new col customer_category according to the given condition

q = '''SELECT first_name || ' ' || last_name Customer_Name, 
        count(inv.invoice_id) No_Of_Invoices,
        sum(total) Total_Spent,
        Case
            when sum(inv.total) < 40 then "Small Spender"
            when sum(inv.total) > 100 then "Big Spender"
            Else "Regular"
        End as Customer_category
        from customer c
        inner join invoice inv
        on 
        c.customer_id = inv.customer_id
        group by c.customer_id
        order by 1
        
'''
sql_query(q)
```

```python
# value_count 
q = '''SELECT first_name || ' ' || last_name Customer_Name, 
        count(inv.invoice_id) No_Of_Invoices,
        sum(total) Total_Spent,
        Case
            when sum(inv.total) < 40 then "Small Spender"
            when sum(inv.total) > 100 then "Big Spender"
            Else "Regular"
        End as Customer_category
        from customer c
        inner join invoice inv
        on 
        c.customer_id = inv.customer_id
        group by c.customer_id
        order by 1
        
'''
sql_query(q)["Customer_category"].value_counts()
```

```python
# Create table

q = """create view CUSTOMER_ABV3_90 as 
        select c.*, sum(inv.total) Total_Spent
        from customer c
        left join invoice inv
        on
        c.customer_id = inv.customer_id
        group by c.customer_id
        having Total_Spent > 90
        order by Total_Spent DESC limit 3
        

"""
conn.execute(q)
```

```python
# Summary data for every palylist in the chinook database

q = """With 
       playlist_info as (
       Select
       pl.playlist_id playlist_id, 
       pl.name playlist_name,
       t.name track_name, 
       (t.milliseconds/1000) length_seconds
       From playlist pl
       Left Join playlist_track plt 
       on pl.playlist_id = plt.playlist_id
       left join track t
       On plt.track_id = t.track_id
       )
       
       Select playlist_id, playlist_name, count(track_name) number_of_tracks, 
       Sum (length_seconds) length_seconds from playlist_info 
       Group BY 1,2
       Order By 1 """

sql_query(q)
```

```python
#  customer from each country that has spent the most money at our store
# Best customer_id, Customer_name, Country, Top_Buyer

q = """with best_customer_id as
       (select c.customer_id, c.first_name ||" "|| c.last_name full_name, c.country,
       sum(i.total) Total_purchased
       from customer c
       inner join invoice i 
       on
       c.customer_id = i.customer_id
       group by 1)
       
       
       select customer_id, country, full_name, Total_purchased
       from best_customer_id


    """
sql_query(q)
```

```python
#  customer from each country that has spent the most money at our store
# Best customer_id, Customer_name, Country, Top_Buyer

q = """with best_customer_id as
       (select c.customer_id, c.first_name ||" "|| c.last_name full_name, c.country,
       sum(i.total) Total_purchased
       from customer c
       inner join invoice i 
       on
       c.customer_id = i.customer_id
       group by 1)
       
       
       select customer_id, country, full_name, max(Total_purchased) Top_buyer
       from best_customer_id
       group by 2


    """
sql_query(q)
```
